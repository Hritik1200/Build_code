```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x = np.array([8,10,12,15,20,25]).reshape(-1,1) # independent variable
y = np.array([10,13,16,20,25,30]) # Dependent Variable
```

```python
# create a train the model

model = LinearRegression()
model.fit(x,y)
```

```python
# get the slope (m) and intercept (b)

m = model.coef_[0] # slope

b = model.intercept_ # intercept

print(f"slope {m}")
print(f"intercept {b}")
```

```python
# predicting price for a 20 inch pizza

x_new = np.array([22.5]).reshape(-1,1)

y_pred  = model.predict(x_new)

print(f"predicted price {y_pred[0]:.2f}")
```

```python
# plot

plt.scatter(x,y,color = "blue",label = "Data point")

plt.plot(x,model.predict(x),color = "green",label = "Regression line")

plt.xlabel("Diameter in inches")
plt.ylabel("pizza price prediction")

plt.show()



```
